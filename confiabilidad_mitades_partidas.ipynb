### confiabilidad Mitades Partidas

Requiere una sola medición al mismo grupo, se divide la prueba en dos mitades y se comparan las correlaciones de los resultados

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Carga el archivo Excel con los resultados simulados
df = pd.read_excel('data/Resultados_Metodo_Mitades_Partidas.xlsx')

# Dividiendo el instrumento en dos mitades
mitad1 = df.iloc[:, :10]  # Primera mitad del test
mitad2 = df.iloc[:, 10:]  # Segunda mitad del test

# Calculando la suma de los puntajes para cada mitad y para cada participante
suma_mitad1 = mitad1.sum(axis=1)
suma_mitad2 = mitad2.sum(axis=1)

# Calculando la correlación de Pearson entre las sumas de las dos mitades
correlacion, p_valor = pearsonr(suma_mitad1, suma_mitad2)

# Imprimiendo la correlación y el valor p
print(f'Correlación de Pearson: {correlacion}')
print(f'Valor p: {p_valor}')

# Generando un gráfico de dispersión para visualizar la relación
plt.figure(figsize=(10, 6))
plt.scatter(suma_mitad1, suma_mitad2, color='blue', alpha=0.5)
plt.title('Correlación entre las Mitades Partidas del Instrumento')
plt.xlabel('Suma de Puntajes - Mitad 1')
plt.ylabel('Suma de Puntajes - Mitad 2')
plt.grid(True)
plt.show()
```

### Interpretación de la Correlación de Pearson
La correlación de Pearson mide la relación lineal entre dos conjuntos de datos, y su valor puede oscilar entre -1 y 1.

Valor cercano a 1: Indica una fuerte correlación positiva, donde un aumento en un conjunto de datos se asocia con un aumento en el otro.
Valor cercano a -1: Significa una fuerte correlación negativa, donde un aumento en un conjunto de datos se asocia con una disminución en el otro.
Valor cercano a 0: Sugiere una falta de correlación lineal significativa entre los conjuntos de datos.
En tu caso, una correlación de -0.0736995088438563 indica una correlación negativa muy débil entre las dos mitades partidas del instrumento. Esto sugiere que no hay una relación lineal significativa entre los puntajes de las dos mitades; es decir, los puntajes de una mitad no predicen de manera confiable los puntajes de la otra mitad.

### Interpretación del Valor p
El valor p te dice si el resultado observado (en este caso, la correlación) es estadísticamente significativo.

Un valor p menor que 0.05 generalmente se considera estadísticamente significativo, indicando que los resultados son poco probables que hayan ocurrido por casualidad.
Un valor p de 0.46617229680638855, como en tu caso, sugiere que no hay suficiente evidencia para considerar que la correlación observada difiere significativamente de 0. En otras palabras, la débil correlación negativa observada podría ser simplemente el resultado del azar.

### Conclusión
La combinación de una correlación cercana a cero y un valor p alto indica que no hay una relación significativa entre las puntuaciones de las dos mitades del test. Esto podría sugerir varios puntos para considerar:

Posible falta de coherencia interna: El instrumento podría no ser consistente consigo mismo a través de las dos mitades, lo que puede indicar problemas con la fiabilidad del instrumento.
Necesidad de revisión del instrumento: Podría ser necesario revisar las preguntas del instrumento para asegurarse de que todas midan el mismo constructo o considerar otros métodos para evaluar la fiabilidad del instrumento.
Este resultado llama a una reflexión más profunda sobre el diseño del instrumento y posiblemente a una revisión para mejorar su consistencia interna y, por ende, su fiabilidad.
