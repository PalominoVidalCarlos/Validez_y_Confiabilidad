### 1. Cargar los datos

```python
import pandas as pd

# Cargar los datos
archivo_excel = 'data/respuestas_encuesta_likert.xlsx'
df = pd.read_excel(archivo_excel)

# Verificar los primeros registros para asegurarse de que los datos se han cargado correctamente
print(df.head())
```

### 2. Verificar valores nulos o faltantes


```python
# Contar valores nulos
valores_nulos = df.isnull().sum()
print(valores_nulos)
```

### 3. Estadísticas descriptivas básicas


```python
# Estadísticas descriptivas
estadisticas_descriptivas = df.describe()
print(estadisticas_descriptivas)

# Media y mediana
media = df.mean()
mediana = df.median()

# Varianza y desviación estándar
varianza = df.var()
desviacion_estandar = df.std()

print("Media:\n", media)
print("Mediana:\n", mediana)
print("Varianza:\n", varianza)
print("Desviación estándar:\n", desviacion_estandar)
```

### 4. Histogramas de las distribuciones de respuestas


```python
import matplotlib.pyplot as plt

# Histogramas para visualizar la distribución de las respuestas
df.hist(bins=5, edgecolor='black', figsize=(14, 10))
plt.tight_layout()
plt.show()
```

### 5. Diagramas de caja para identificar valores atípicos


```python
# Diagramas de caja para cada pregunta
df.plot(kind='box', figsize=(14, 8), vert=False)
plt.title('Diagramas de Caja para Cada Pregunta')
plt.show()
```

### 6. Correlación entre preguntas


```python
import seaborn as sns

# Mapa de calor de la correlación
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlación entre Preguntas')
plt.show()
```

### 7. Detección de valores atípicos (opcional, avanzado)

Podrías querer identificar los valores atípicos utilizando el método de la distancia intercuartílica (IQR). Esto es más avanzado y opcional dependiendo de tu necesidad de limpiar los datos.


```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Identificar valores atípicos
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print(outliers)
```
